```python
# Install: pip install diffusers transformers accelerate safetensors torch torchvision
from diffusers import StableDiffusionImg2ImgPipeline
import torch
from PIL import Image
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model_id = "runwayml/stable-diffusion-v1-5"  # example base model
```

```python
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(model_id, torch_dtype=torch.float16 if device=="cuda" else torch.float32)
pipe = pipe.to(device)
```

```python
pipe.safety_checker = lambda images, **kwargs: (images, [False])
```

```python

# load sprite (RGBA) — you may want to convert/resize
init_image = Image.open("pokemon_imgs/3.png").convert("RGBA").resize((96,96))
```

```python
prompt = "A fantasy creature, highly detailed, full body, studio lighting, colorful, concept art"
negative_prompt = "watermark, text, lowres"
```

```python
# strength ∈ (0.0, 1.0) — lower keeps more of original image
out = pipe(prompt=prompt, image=init_image, strength=0.6, guidance_scale=2, num_inference_steps=3)
out.images[0].save("img2img_output.png")
```
